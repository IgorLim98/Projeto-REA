# Sistema de equaes lineares  concentrao de reatores

Sistema de equaes so vistos com frequncia em problemas que envolvem a engenharia qumica. Com o princpio de conservao das massas, vrios problemas so modelados utilizando essas entidades matemticas. Talvez o exemplo mais famoso  o que envolve vrios reatores ligados entre si. O princpio de conservao de massas resulta em balanos efetuados em um recipiente (no caso um reator) expressos matematicamente por:

<h3 align="center">Acumulao = Entrada  Sadas</h3> 



Caso as entradas sejam maiores que sadas, tm  se um aumento da massa no interior, caso contrrio, as sadas sejam maiores que as entradas, a massa no interior diminui. Chamamos de estado estacionrio, ou regime permanente, quando o somatrio das entradas se iguala ao somatrio das sadas. Dessa forma o acmulo  zerado e sobra:

<h3 align="center">Entrada = Sadas</h3>

Imagine a seguinte situao. 5 reatores em srie conectados entre sim, nos quais a alimentao de um ou mais reatores  a sada de um ou mais reatores, como mostra a imagem a seguir:



Na qual $Q_{ij}$ (vazo) representa a sada do reator $i$ alimentando o reator $j$. Por exemplo, $Q_{24}$ informa que uma das sadas do reator 2 alimenta o reator 4. As concentraes de cada reator so denotadas pelos $c_{i}$ ($i = 1, 2, ..., 5$). A taxa de massa que sai (ou entra) em cada reator  calculada multiplicando a vazo, dada em  $\large \frac{m^3}{min}$ , com a concentrao dada em $\large \frac{mg}{m^3}$. 

Para o reator 1, h duas entradas e duas sadas como indicam as setas na figura acima, ento o balano de massa fica:

$$\large Q_{01} c_{01}  + Q_{31} c_{3}  = Q_{15} c_{1}+ Q_{12} c_{1}$$

Vamos aplicar mtodo numrico para resolver situao  problema semelhante da figura acima, onde os valores das vazes so fornecidos, como as concentraes que alimentam os reatores 1 e 3. O objetivo  determinar as concentraes dos 5 reatores ($c1, c2, c3, c4, c5$). 
As razes fornecidas so: 

$$\large Q_{01} = 5, Q_{31} = 3, Q_{25} = 2, Q_{23} = 2$$

$$\large Q_{15} = 4, Q_{55} = 3, Q_{54} = 3, Q_{34} = 7$$

$$\large Q_{12} = 4, Q_{03} = 8, Q_{24} = 0, Q_{44} = 10, c_{01} = 10, c_{03} = 20$$

Repetindo o procedimento acima, para os 5 reatores, temos o seguinte sistema:

$$\large 8c_{1}  - 3c_{3}  = 50$$

$$\large 4c_{1}  - 4c_{2}  = 0$$

$$\large 10c_{3}  - 2c_{2}  = 160$$

$$\large -7c_{3} + 10c_{4} - 3c_{5}  = 0$$

$$\large 4c_{1}  +2c_{2}  -6c_{5}  = 0$$

O sistema acima pode ser reescrito na forma de multiplicao matricial, do tipo:

$$\large Ax + b = 0$$

Onde A, no nosso caso, representa a matriz 5x5 dos coeficientes do sistema
acima, x  o vetor coluna 5x1 com os valores das concentraes que queremos determinar, por ltimo, b  vetor coluna dos componentes independentes.

$$\large A = \begin{bmatrix} 8 & 0 & -3 & 0 & 0 \\ 4 & -4 & 0 & 0 & 0 \\ 0 & -2 & 10 & 0 & 0 \\ 0 & 0 & -7 & 10 & -3 \\ 4 & 2 & 0 & 0 & -6 \end{bmatrix}$$

$$\large b = \begin{bmatrix} -50 \\ 0 \\ -160 \\ 0 \\ 0 \end{bmatrix}$$

$$\large x = \begin{bmatrix} C_{1} \\ C_{2} \\ C_{3} \\ C_{4} \\ C_{5} \end{bmatrix}$$

Utilizando o Linsolve, funo embutida para o Scilab, podemos resolver o sistema acima e encontrar os valores das concentraes. A funo linsolve calcula todas as solues para problemas do tipo $Ax + b = 0$.

```octave
// sistema de equaes algbricas lineares
// utilizaremos a funo Linsolve para resolver um sistema simulando a 5 reatores ligados entre si
// Queremos determinar as concetraes c1, c2, c3, c4, c5 (referente a cada reator)

clear, clc

//informe aqui a seu sistema do tipo Ax + b = 0
A = [8 0 -3 0 0;4 -4 0 0 0;0 -2 10 0 0; 0 0 -7 10 -3;4 2 0 0 -6] // matriz dos coeficientes
b = [-50; 0; -160; 0; 0] //vetor coluna dos termos independentes
x = linsolve(A, b)

printf("O vetor x : \n")
mprintf("C1 = %6.4f \n", x(1))
mprintf("C2 = %6.4f \n", x(2))
mprintf("C3 = %6.4f \n", x(3))
mprintf("C4 = %6.4f \n", x(4))
mprintf("C5 = %6.4f \n", x(5))
```

Ento, aplicando o cdigo no Scilab, temos que o vetor $x$  igual a:

$$\large x = \begin{bmatrix} c_{1} = 13.2432 \\ c_{2} = 13.2432 \\ c_{3} = 18.6486 \\ c_{4} = 17.0270 \\ c_{5} = 13.2432 \end{bmatrix}$$
